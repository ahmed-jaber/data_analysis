```python
import pandas as pd

df=pd.read_csv('failures.csv')

df.info()
```

```python
import matplotlib.pyplot as plt

xPoints=df['Age']
yPoints=df['Failures']

plt.xlabel('Age')
plt.ylabel('Failures')

plt.plot(xPoints,yPoints)
```

```python
xPoints=df['Age']
y1Points=df['Failures']
y2Points=df['DaysOff']

plt.subplot(1,2,1)
plt.xlabel('Age')
plt.ylabel('Failures')
plt.scatter(xPoints,y1Points)

plt.subplot(1,2,2)
plt.xlabel('Age')
plt.ylabel('DaysOff')
plt.scatter(xPoints,y2Points)
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df)
```

```python
df=pd.read_csv('stdMarks.csv')
dfCorr=df[['Math','Phy','Lang']]

correlations=dfCorr.corr(method='pearson')
print(correlations)

fig=plt.figure()

ax=fig.add_subplot(1,1,1)

cax=ax.matshow(correlations)

ticks=[0,1,2]

ax.set_xticks(ticks)
ax.set_yticks(ticks)

ax.set_xticklabels(['Math','Phy','Lang'])
ax.set_yticklabels(['Math','Phy','Lang'])
```
